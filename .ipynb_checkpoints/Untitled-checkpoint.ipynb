```python
import pandas as pd
import numpy as np
import pymysql
from sqlalchemy import create_engine
from sqlalchemy import text
import getpass  # To get the password without showing the input
password = getpass.getpass()
```

```python
sd = "sakila"
connection_string = 'mysql+pymysql://root:' + password + '@localhost/'+sd
engine = create_engine(connection_string)
engine
```

```python
def rentals_month(engine, month, year):
    with engine.connect() as connection:
        query = text(f'''select rental_id, rental_date, inventory_id, customer_id, return_date, staff_id 
                         from rental
                         where month(rental_date) = {month} and year(rental_date) = {year}
                         order by rental_date asc;''')
        
        result = connection.execute(query)
        return pd.DataFrame(result.fetchall())

rentals_df = rentals_month(engine, 5, 2005)

rentals_df
```

```python
def rental_count_month(rentals_df, month, year):
    rental_counts = rentals_df.groupby("customer_id")["rental_id"].count().reset_index()
    rental_counts.rename(columns={"rental_id": f"rentals_{month}_{year}"}, inplace=True)
    rentals_df=rentals_df.drop(columns=['rental_id', 'inventory_id', 'staff_id', 'return_date'])

    return rental_counts

rental_counts_df = rental_count_month(rentals_df,5,2005)

rental_counts_df
```

```python
def compare_rentals(rentals_df, rental_counts_df):
    df_merge=pd.merge(rentals_df, rental_counts_df, on = "customer_id")
    df_merge['difference']=df_merge.iloc[:,0]-df_merge.iloc[:,1]

    return df_merge
```

```python
rentals_df = rental_count_month(rentals_df,5,2005)
rental_counts_df = rental_count_month(rentals_df,5,2005)

compare_rentals(rentals_df, rental_counts_df)
```
